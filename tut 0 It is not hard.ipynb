# simple display of what you can do with very little code!
displaying population of each african countries, where size scale corresponds to the population size

```python
# import alll packages we need
import geoplot
import geopandas
```

```python
world = geopandas.read_file(
    geopandas.datasets.get_path('naturalearth_lowres')
)
```

```python
geoplot.polyplot(world, figsize=(8, 4))
```

```python
ax = geoplot.polyplot(
    world, projection=geoplot.crs.Orthographic(), figsize=(8, 4)
)
# if you want to see outline of the earth as well: 
#ax.outline_patch.set_visible(True)
```

limits: so that the smallest point is not too small --- some country has very small population, we don't want it to disapear!

```python
africa = world.query('continent == "Africa"')
ax = geoplot.cartogram(
    africa, scale='pop_est', limits=(0.1, 1),
    edgecolor='None', figsize=(7, 8), legend=True, 
    color='purple',
)
geoplot.polyplot(africa, edgecolor='grey', ax=ax)
```

# homework: can you imagine how to make the same plot with another continent? Maybe play with projection type?

```python
africa = world.query('continent == "Asia"')
ax = geoplot.cartogram(
    africa, scale='pop_est', limits=(0.1, 1),
    edgecolor='None', figsize=(7, 8), legend=True, 
    color='purple',
    projection=geoplot.crs.Orthographic()
)
geoplot.polyplot(africa, edgecolor='grey', ax=ax)
```
